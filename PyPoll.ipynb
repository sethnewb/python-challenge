```python
# import os library
# import csv library
import os
import csv

# set the file location as a variable
csvpath = "C:/Users/sethn/Desktop/python-challenge/election_data.csv"

# create the list variable
VtrData = []

# create csv from file path and import into list
with open(csvpath, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    print(csvreader)
    csv_header = next(csvreader)
    VtrData.append(csv_header)
    for row in csvreader:
        VtrData.append(row)
        
# Set a variable and find the count of rows
# Number of rows is equal to how many votes have been cast
RwCnt = 0
for row in VtrData[1:]:
    RwCnt += 1

# Create a dynamic library to store candidate names and values
# for every list besides the header in voter data
#    if the candidate name is not in the dictionary
#        Add the candidate name as a dictionary
#        Add a vote key and assigned it a value of 1
#        Add a percent key and set it equal to votes divided by the total votes
#    if the candidate name is in the list
#        Add 1 to the Votes key
#        calculate new voter percentage
Candidates = {}
for row in VtrData[1:]:
    if row[2] not in Candidates:
        Candidates[row[2]] = {}
        Candidates[row[2]]["Votes"] = 1
        Candidates[row[2]]["Prcnt"] = (Candidates[row[2]]["Votes"] / RwCnt)
    else:
        Candidates[row[2]]["Votes"] += 1
        Candidates[row[2]]["Prcnt"] = (Candidates[row[2]]["Votes"] / RwCnt)

# Create variables to hold the winner name and vote count
# for each name dictionary in the Candidates dictionary
#    if the votes key of the current dictionary is greater than the
#     -currently held most votes
#        Update the most votes variable to the votes key value of the 
#         -current dictionary
#        Update the winner name to the current name dictionary
MstVts = 0
MstVtsNm = ""
for names in Candidates:
    if Candidates[names]["Votes"] > MstVts:
        MstVts = Candidates[names]["Votes"]
        MstVtsNm = names
        
print("Election Results")
print("--------------------------")
print(f"Total Votes:  {RwCnt}")
print("--------------------------")
for key in Candidates:
    print(key + ": " + str(format(Candidates[key]["Prcnt"], ".3%")) + " (" + str(Candidates[key]["Votes"]) + ")")
print("--------------------------")
print(f"Winner: {MstVtsNm}")
print("--------------------------")


with open("PollAnalysis.txt", "a") as txt:
    txt.write("Election Results\n")
    txt.write("--------------------------\n")
    txt.write(f"Total Votes:  {RwCnt}\n")
    txt.write("--------------------------\n")
    for key in Candidates:
        txt.write(key + ": " + str(format(Candidates[key]["Prcnt"], ".3%")) + " (" + str(Candidates[key]["Votes"]) + ")\n")
    txt.write("--------------------------\n")
    txt.write(f"Winner: {MstVtsNm}\n")
    txt.write("--------------------------\n")
txt.close()
```
